The following function is an implementation of binary search shown in the lecture.
- Introduce a variable `count` to count the number of comparisons between `target` and the elements in the list.
- Modify the function to return a tuple of the index and the number of comparisons if the target is found, and a tuple of -1 and the number of comparisons if the target is not found.
- Verify the function's behavior.

```python
def binary_search(arr:list, target) ->int:
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```
